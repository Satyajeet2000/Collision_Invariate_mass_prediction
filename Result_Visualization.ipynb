```python
import matplotlib.pyplot as plt

# Data for MSE
model_names = ['Decision Tree', 'Random Forest', 'LightGBM', 'CatBoost', 'RNN', 'Deep FeedForward Network']
mse_values = [148.326, 35.74, 28.147, 4.994, 3.532, 2.652]

# Plotting MSE
plt.figure(figsize=(12, 6))
plt.plot(model_names, mse_values, label='Mean Squared Error', color='blue', marker='o', markersize=8)
plt.title('Mean Squared Error of Different Models', fontsize=16)
plt.ylabel('Mean Squared Error', fontsize=12)
plt.xticks(range(len(model_names)), model_names, rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for R2 Score
r2_scores = [0.765, 0.943, 0.955, 0.992, 0.994, 0.995]

# Plotting R2 Score
plt.figure(figsize=(12, 6))
plt.plot(model_names, r2_scores, label='R2 Score', color='red', marker='s', markersize=8)
plt.title('R2 Score of Different Models', fontsize=16)
plt.ylabel('R2 Score', fontsize=12)
plt.xticks(range(len(model_names)), model_names, rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
