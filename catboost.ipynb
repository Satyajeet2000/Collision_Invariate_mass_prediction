```python
import pandas as pd
from catboost import CatBoostRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
data = pd.read_csv("Dataset.csv")
```

```python
x = data.drop("M",axis=1)
y = data["M"]

scaler = StandardScaler()
x = scaler.fit_transform(x)

X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
catbrmodel = CatBoostRegressor().fit(X_train, Y_train)
```

```python
catbrmodel_pred =catbrmodel.predict(X_test)

print("Mean Squared Error:", mean_squared_error(Y_test, catbrmodel_pred))
print("R2 Score:", r2_score(Y_test, catbrmodel_pred))
```
