```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Preprocessed_data.csv")
```

```python
fig,ax = plt.subplots(nrows=7,ncols=1,figsize=(16,35))

sns.scatterplot(ax= ax[0],data=data, x='pt1', y='pt2', hue='M', hue_norm=(data.M.min(),data.M.max()))
ax[0].set_title('Mass by linear momentum')
sns.scatterplot(ax= ax[1],data=data, x='eta1', y='eta2', hue='M', hue_norm=(data.M.min(),data.M.max()))
ax[1].set_title('Mass by pseudorapidity')
sns.scatterplot(ax= ax[2],data=data, x='E1', y='E2', hue='M', hue_norm=(data.M.min(),data.M.max()))
ax[2].set_title('Mass by energy')
sns.scatterplot(ax= ax[3],data=data, x='phi1', y='phi2', hue='M', hue_norm=(data.M.min(),data.M.max()))
ax[3].set_title('Mass by phi angle')
sns.scatterplot(ax= ax[4],data=data, x='px1', y='px2', hue='M', hue_norm=(data.M.min(),data.M.max()))
ax[3].set_title('Mass by momentum of x')
sns.scatterplot(ax= ax[5],data=data, x='py1', y='py2', hue='M', hue_norm=(data.M.min(),data.M.max()))
ax[3].set_title('Mass by momentum of y')
sns.scatterplot(ax= ax[6],data=data, x='pz1', y='pz2', hue='M', hue_norm=(data.M.min(),data.M.max()))
ax[3].set_title('Mass by momentum of z')

plt.show()
```

```python
plt.figure(figsize=(16,9))
heatmap = sns.heatmap(data.corr(), vmin=-1,vmax=1, annot=True)

heatmap.set_title('Correlation Heatmap')
plt.show()
```
